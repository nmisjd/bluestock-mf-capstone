```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (10, 5)
sns.set_style("whitegrid")
```

```python
nav_df = pd.read_csv("../data/processed/clean_nav_history.csv")

perf_df = pd.read_csv(
    "../data/processed/clean_scheme_performance.csv"
)

tx_df = pd.read_csv(
    "../data/processed/clean_transactions.csv"
)

aum_df = pd.read_csv(
    "../data/raw/03_aum_by_fund_house.csv"
)

sip_df = pd.read_csv(
    "../data/raw/04_monthly_sip_inflows.csv"
)
```

```python
nav_df["date"] = pd.to_datetime(nav_df["date"])

tx_df["transaction_date"] = pd.to_datetime(
    tx_df["transaction_date"]
)

aum_df["date"] = pd.to_datetime(aum_df["date"])
```

```python
print(nav_df.shape)
print(perf_df.shape)
print(tx_df.shape)
print(aum_df.shape)
print(sip_df.shape)
```

```python
sample_fund = nav_df[
    nav_df["amfi_code"] == 119551
]

plt.plot(
    sample_fund["date"],
    sample_fund["nav"]
)

plt.title("SBI Bluechip NAV Trend")
plt.xlabel("Date")
plt.ylabel("NAV")

plt.show()
```

```python
latest_aum = aum_df.sort_values(
    by="aum_crore",
    ascending=False
)

sns.barplot(
    data=latest_aum.head(10),
    x="aum_crore",
    y="fund_house"
)

plt.title("Top Fund Houses by AUM")

plt.show()
```

```python
plt.plot(
    sip_df["month"],
    sip_df["sip_inflow_crore"]
)

plt.xticks(rotation=90)

plt.title("Monthly SIP Inflows Trend")
plt.xlabel("Month")
plt.ylabel("SIP Inflow (Crore)")

plt.show()
```

```python
sns.scatterplot(
    data=perf_df,
    x="std_dev_ann_pct",
    y="return_3yr_pct",
    hue="risk_grade"
)

plt.title("Risk vs Return Analysis")

plt.show()
```

```python
tier_counts = tx_df["city_tier"].value_counts()

plt.pie(
    tier_counts,
    labels=tier_counts.index,
    autopct="%1.1f%%"
)

plt.title("Transactions by City Tier")

plt.show()
```

```python
corr_cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "alpha",
    "beta",
    "sharpe_ratio",
    "std_dev_ann_pct"
]

corr_matrix = perf_df[corr_cols].corr()

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Performance Metrics Correlation")

plt.show()
```
